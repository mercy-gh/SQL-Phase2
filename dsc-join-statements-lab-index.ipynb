```python
# Your code here
pd.read_sql(""" SELECT firstName, lastName,productName 
            FROM employees as e
            JOIN
            customers as c on e.employeeNumber=c.salesRepEmployeeNumber
            JOIN
            orders as o USING(customerNumber)
            JOIN
            orderdetails AS od USING(orderNumber)
            JOIN
            products USING(productCode)
            GROUP BY employeeNumber
            """,conn)
```

## Level Up 3: Display the names employees who have sold more than 200 different products

Hint: this is different from the previous question because the quantity sold doesn't matter, only the number of different products

```python
# Your code here
pd.read_sql(""" SELECT employeeNumber, firstName, lastName, productName, COUNT(productCode)
            FROM employees as e
            JOIN
            customers as c on e.employeeNumber=c.salesRepEmployeeNumber
            JOIN
            orders as o USING(customerNumber)
            JOIN
            orderdetails AS od USING(orderNumber)
            JOIN
            products USING(productCode)
            GROUP BY employeeNumber
            HAVING COUNT(productCode)>200
            """,conn)
```

```python
#Select names of employees from employees table who have sold more than 200 different products
# joining tables: employees, orders, orderdetails and product,
# 
pd.read_sql(""" SELECT employeeNumber, firstName, lastName, productName, COUNT(productCode)
            FROM employees as e
            JOIN
            customers as c on e.employeeNumber=c.salesRepEmployeeNumber
            JOIN
            orders as o USING(customerNumber)
            JOIN
            orderdetails AS od USING(orderNumber)
            JOIN
            products USING(productCode)
            GROUP BY employeeNumber
            HAVING COUNT(DISTINCT(productCode))>200
            """,conn)
```
